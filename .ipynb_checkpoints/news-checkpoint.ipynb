```python
import requests
import bs4
from pprint import pprint  #อันนี้ต้องลบออก ใข้ไม่ได้

import pandas as pd
import numpy as np
# from pandas import DataFrame
```

```python
links = []
response = requests.get('http://www.it.kmitl.ac.th/~teerapong/news_archive/index.html')
html_page = bs4.BeautifulSoup(response.content)
selector = 'li > a'
tags = html_page.select(selector)
for tag in tags:
    link = 'http://www.it.kmitl.ac.th/~teerapong/news_archive/' + tag['href']
    print(link)
    links.append(link)
```

```python
'''
This is a multiline
comment.
'''
category_news = []
for link in links:
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'tbody > tr > td.category'
    tags = html_page.select(selector)
    #pprint(tags)
    #tags[0].text.strip()
    #len(tags) # หาจำนวนข่าวในเดือนมกรา
    for i in range(len(tags)):
        category = tags[i].text.strip()
        print(category)
        category_news.append(category)
```

```python
# url = 'http://www.it.kmitl.ac.th/~teerapong/news_archive/month-jan-2017.html' 
# response = requests.get(url)
# html_page = bs4.BeautifulSoup(response.content)
# print(html_page)
```

```python
df = pd.DataFrame(category_news,columns=['Category_name'])
df
```

```python
df = df.assign(code = category_news)
df
```

```python
df.code = pd.Categorical(df.code)
```

```python
df['code'].dtypes
```

```python
df['code'] = df['code'].cat.codes
df
```

```python
df.to_csv("category_news.csv")
```

## ดึงข้อมูลข่าว

```python
links_article = []
for link in links:
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'tbody > tr > td.title > a'
    tags = html_page.select(selector)

    for tag in tags:
        link = 'http://www.it.kmitl.ac.th/~teerapong/news_archive/' + tag['href']
        print(link)
        links_article.append(link)
```

```python
pprint(links_article)
```

```python
titles = []
for link in links_article:
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'h2'
    tags = html_page.select(selector)

    title = tags[0].text
    print(title)
    titles.append(title)
```

```python
pprint(titles)
```

```python
#ต้องไปถามอาจารย์ว่าใช้ได้ไหม
import time
```

```python
text = ''
articles_only = []
links_article = ['http://www.it.kmitl.ac.th/~teerapong/news_archive/article-jan-0418.html', 'http://www.it.kmitl.ac.th/~teerapong/news_archive/article-jan-0027.html']
for link in links_article:
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'p'
    tags = html_page.select(selector)

    for i in range(len(tags)-1):
        if tags[i].string is None:
            continue
        else:
            paragraph = tags[i].string.replace(u'\n', u' ').replace("\'s", "'s")
            text += paragraph
    articles_only.append(text)
```

```python
articles_only
```

```python
articles_dataset = '\n'.join(articles_only)
```
