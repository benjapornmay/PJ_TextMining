### Import Packages Etc.

```python
import requests
import bs4
from pprint import pprint  #อันนี้ต้องลบออก ใข้ไม่ได้

import pandas as pd
import numpy as np
# from pandas import DataFrame
```

## Part 1: Data Collection

```python
#เก็บ link 12 เดือน
links = []
response = requests.get('http://www.it.kmitl.ac.th/~teerapong/news_archive/index.html')
html_page = bs4.BeautifulSoup(response.content)
selector = 'li > a'
tags = html_page.select(selector)
for tag in tags:
    link = 'http://www.it.kmitl.ac.th/~teerapong/news_archive/' + tag['href']
    print(link)
    links.append(link)
```

```python
'''
เก็บcategory
'''

category_news = []

for link in links:
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'tbody > tr > td.category'
    tags = html_page.select(selector)
    
    for catagory_name in range(len(tags)):
        if catagory_name != 'N/A':
            category = tags[catagory_name].text.strip()
            category_news.append(category)
```

```python
df = pd.DataFrame(category_news, columns=['Category_name'])
df
```

```python
df = df.assign(code = category_news)
df
```

```python
df.code = pd.Categorical(df.code)
```

```python
df['code'].dtypes
```

```python
df['code'] = df['code'].cat.codes
df
```

```python
df.to_csv("category_news.csv")
```

## ดึงข้อมูลข่าว

```python
links_article = []
for link in links:
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'tbody > tr > td.title > a'
    tags = html_page.select(selector)

    for tag in tags:
        link = 'http://www.it.kmitl.ac.th/~teerapong/news_archive/' + tag['href']
        print(link)
        links_article.append(link)
```

```python
titles = []
for link in links_article:
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'h2'
    tags = html_page.select(selector)

    title = tags[0].text
    titles.append(title)
```

```python
pprint(titles)
```

```python
articles_only = []
count = 0

for link in links_article:
    text = ''
    response = requests.get(link)
    html_page = bs4.BeautifulSoup(response.content)
    selector = 'p'
    tags = html_page.select(selector)
    for i in range(len(tags)):
        if i == (len(tags) - 1):
            break
        else:
            if tags[i].string is None:
                continue
            else:
                paragraph = tags[i].string.replace('\n', ' ')
                text += paragraph
    time.sleep(1)        
    articles_only.append(text)

    count+=1
    print(count, text)
```

```python
articles_dataset = '\n'.join(articles_only)
```
